## 1. Molecules move

```python
transitions = np.array([12399.8, 9826.48,7149.04,4367.50,1481.86])
levels = np.array([0.5,1.5,2.5,3.5,4.5])
```

## 2. How can we quantify the movement

```python
from scipy.stats import linregress
fit = linregress(levels, transitions)
```

```python
x_max = -fit[1]/fit[0]
new_x = np.linspace(0,x_max, 20)
new_y = fit[0]*new_x+fit[1]
```

## 3. How can we measure the movement

```python
plt.plot(new_x,new_y)
```

```python
A = 0.5*max(new_y)*max(new_x)
```
